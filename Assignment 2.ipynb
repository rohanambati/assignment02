```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

df3 = pd.read_csv('/content/drive/MyDrive/housing.csv')
```

```python
features = df3.drop('ocean_proximity', axis=1)
target = df3['median_house_value']
```

```python
scaler = MinMaxScaler()
features_normalized = scaler.fit_transform(features)
df_normalized = pd.DataFrame(features_normalized, columns=features.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_normalized, target, test_size=0.2, random_state=42)
```

```python
print("Train set shape:", X_train.shape, y_train.shape)
print("Test set shape:", X_test.shape, y_test.shape)
```
